### 실습 2-2 : 시퀀스 원소의 최댓값 출력하기

```python
from typing import Any, Sequence

def max_of(a: Sequence)-> Any:
    #시퀀스형 a 원소의 최댓값을 반환
    maximum = a[0]
    for i in range(1, len(a)):
        if(a[i] > maximum):
            maximum = a[i]
    
    return maximum

if __name__ == '__main__':
    print('배열을 최댓값을 구합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    
    x = [None]* num
    
    for i in range(num):
        x[i]= int(input(f'x[{i}]값을 입력하세요: '))
    
    print(f'최댓값은 {max_of(x)}입니다.')
```

### 실습 2-3: 배열 원소의 최댓값을 구해서 출력하기 (원소값 입력 받음)

```python
#from max import max_of

print('배열의 최댓값을 구합니다.')
print('주의: "End"를 입력하면 종료합니다.')

number = 0
x = []

while True:
    s = input(f'x[{number}]값을 입력하세요: ')
    if s == 'End':
        break
    else:
        x.append(int(s)) #배열 맨 끝에 추가
        number += 1

print(f'{number}개 입력했습니다.')
print(f'최댓값은 {max_of(x)} 입니다.')
```

### 실습 2-6 : 뮤터블 시퀀스 원소를 역순으로 정렬

```python
from typing import Any, MutableSequence

def reverse_array(a: MutableSequence) -> None:
    #뮤터블 시퀀스 a의 원소를 역순으로 정렬
    n = len(a)
    for i in range(n // 2):
        a[i], a[n-i-1] = a[n-i-1], a[i]

if __name__ == '__main__' :
    
    print('배열 원소를 역순으로 정렬합니다.')
    num = int(input('원소 수를 입력하세요'))
    arr = [None] * num
    
    for i in range(num):
        arr[i] = int(input(f'arr[{i}]값을 입력하세요.: '))
    
    reverse_array(arr)
    
    print('배열 원소를 역순으로 정렬했습니다.')
    for i in range(num):
        print(f'arr[{i}] = {arr[i]}')
```

### 실습 2-7 : 기수변환 (10진수 정수값을 입력받아 2~36진수로 변환하여 출력하기)

```python
def card_conv(x: int, r:int) -> str:
    #정수값 x를 r진수로 변환한 뒤 그 수를 나타내는 문자열 반환
    
    string = '' #변환 후 문자열

    stringChar = '0123456789ABCDEFHIJKLMNOPQRSTUVWXYZ'
    
    while x > 0:
        string += stringChar[x%r] # 해당하는 문자를 꺼내 결합
        x //= r
    
    return string[::-1]

if __name__ == '__main__':
    
    print('10진수를 n진수로 변환합니다.')
    
    while True:
        while True: 
            num = int(input('변환할 값으로 음이 아닌 정수를 입력하세요.: '))
            if num > 0: 
                break
    
        while True:
            cd = int(input('어떤 진수로 변환할까요?: '))
            if 2<= cd <=36 :
                break
        
        print(f'{cd}진수로는 {card_conv(num,cd)}입니다.')
    
        retry = input('한 번 더 변환할까요?(Y ... 예 / N ... 아니요): ')
        if retry in {'N', 'n'}:
            break
```

    : 함수에서 매개변수의 값을 변경하면 데이터 타입에 따라 구분되어 처리
    
    1) 인수가 이뮤터블일 때: 다른 객체를 생성해 그 객체에 대한 참조로 업데이트

    2) 인수가 뮤터블 : 객체 자체를 업데이트, 매개변수값을 변경하면 호출하는 쪽의 실제 인수 값 변경

### 실습 2-8: 1000이하의 소수 나열하기 

```python
counter = 0

for n in range (2, 10001):
    for i in range(2, n):
        counter += 1
        if n % i == 0:
            break
    else: 
        print(n, end=' ')

print(f'나눗셈을 실행한 횟수: {counter}')
```

### 실습 2-9 : 1000이하의 소수 나열하기 (알고리즘 개선1)

```python
counter = 0
ptr = 0
prime = [None]*500

prime[ptr] = 2
ptr +=1

#홀수 대상 소수 나열
for n in range(3, 1001, 2):
    for i in range(1, ptr):
        counter +=1
        if n % prime[i] == 0:
            break
    
    else:
        prime[ptr] = n
        ptr += 1

for i in range(ptr):
    print(prime[i], end=' ')

print()
print(f'나눗셈을 실행한 횟수 : {counter}')
```

### 실습 2-10 : 1000이하의 소수 나열하기 (알고리즘 개선2)

```python
counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1

prime[ptr] = 3
ptr += 1

for n in range(5, 1001, 2):
    i = 1
    while prime[i] * prime[i] <=2 : 
        counter += 2
        if n % prime[i] == 0:
            break
    else:
        prime[ptr] = n
        ptr += 1
        counter += 1

for i in range(ptr):
    print(prime[i], end=' ')

print()
print(f'곱셈과 나눗셈을 실행한 횟수: {counter}')
```
